# 연월일 구하기
연월일 순으로 구성된 8자리의 날짜가 입력으로 주어진다.
 


해당 날짜의 유효성을 판단한 후, 날짜가 유효하다면

”YYYY/MM/DD”형식으로 출력하고,

날짜가 유효하지 않을 경우, -1 을 출력하는 프로그램을 작성하라.


연월일로 구성된 입력에서 월은 1~12 사이 값을 가져야 하며

1일 ~ 각각의 달에 해당하는 날짜까지의 값을 가질 수 있다.
 
※ 2월의 경우, 28일인 경우만 고려한다. (윤년은 고려하지 않는다.)


[입력]

입력은 첫 줄에 총 테스트 케이스의 개수 T가 온다.

다음 줄부터 각 테스트 케이스가 주어진다.
```
입력
5
22220228
20150002
01010101
20140230
11111111
```

[출력]

테스트 케이스 t에 대한 결과는 “#t”을 찍고, 한 칸 띄고, 정답을 출력한다.

(t는 테스트 케이스의 번호를 의미하며 1부터 시작한다.)
```
출력
#1 2222/02/28
#2 -1
#3 0101/01/01
#4 -1
#5 1111/11/11
```

````python
T = int(input())
for test_case in range(1, T + 1):
    date_num = list(map(int, input()))
    year = str(date_num[0]) + str(date_num[1]) + str(date_num[2]) + str(date_num[3])
    month  = str(date_num[4]) + str(date_num[5])
    day = str(date_num[6]) + str(date_num[7])
    if int(month) < 1 or int(month) > 12:
        result =  -1
    elif (int(month) == 2) & (int(day)<1 or int(day)>28):
        result = -1
    elif (int(month)%2 == 1) & (int(day)<1 or int(day)>31):
        result = -1
    elif (int(month)%2 == 0) & (int(day)<1 or int(day)>30):
        result = -1
    else:
        result = f"{year}/{month}/{day}"
    print(f"#{test_case} {result}")
     
````

# 패턴 마디의 길이
패턴에서 반복되는 부분을 마디라고 부른다. 문자열을 입력 받아 마디의 길이를 출력하는 프로그램을 작성하라.


[제약 사항]

각 문자열의 길이는 30이다. 마디의 최대 길이는 10이다.


[입력]

가장 첫 줄에는 테스트 케이스의 개수 T가 주어지고, 그 아래로 각 테스트 케이스가 주어진다.

각 테스트 케이스의 첫 번째 줄에는 길이가 30인 문자열이 주어진다.
```
입력
3       
KOREAKOREAKOREAKOREAKOREAKOREA
SAMSUNGSAMSUNGSAMSUNGSAMSUNGSA
GALAXYGALAXYGALAXYGALAXYGALAXY    
```

[출력]

출력의 각 줄은 '#t'로 시작하고, 공백을 한 칸 둔 다음 정답을 출력한다.

(t는 테스트 케이스의 번호를 의미하며 1부터 시작한다.)
```
출력
#1 5
#2 7
#3 6
```

````python
T = int(input())
for test_case in range(1, T + 1):
    i = 0
    word_list = list(map(str, input()))
    while True:
        i += 1
        cut_list = word_list[:i]
        cut_num = int(len(word_list) / len(cut_list))+1
        rsmb_list = cut_list * cut_num
        if rsmb_list[:30] == word_list:
            break
    print(f"#{test_case} {len(cut_list)}")
````
